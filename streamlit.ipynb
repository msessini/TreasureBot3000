```python
import streamlit as st
import random
def add_bg_from_local():
    st.markdown(
        f"""
        <style>
        .stApp {{
            background-image: url("./TreasureBot3000/treasurebot/streamlit/robotpic.webp");
            background-size: cover;
            background-position: center;
        }}
        </style>
        """,
        unsafe_allow_html=True
    )
add_bg_from_local()
st.title("TreasureBot 3000")
st.write("Upload an image and click 'Predict' to see in which bin it needs to go.")
uploaded_file = st.file_uploader("Choose an image...", type=["jpg", "jpeg", "png"])
if uploaded_file is not None:
    st.image(uploaded_file)
    if st.button("Predict"):
        # Simulate prediction results
        possible_bins = ["Recyclable", "Non-recyclable", "Compost", "unknown"]
        predicted_bin = random.choice(possible_bins)
        confidence = round(random.uniform(0.5, 1.0), 2)
        st.write("Predictions:")
        if predicted_bin == 'unknown':
            st.warning("This doesn't go in either of the bins!")
        else:
            st.success(f"Bin: {predicted_bin}, Confidence: {confidence}")
sentiment_mapping = ["one", "two", "three", "four", "five"]
selected = st.select_slider("Rate your experience", options=range(5), format_func=lambda x: ":star:" * (x + 1))
if selected is not None:
    st.markdown(f"You selected {sentiment_mapping[selected]} star(s).")
```
